Задача 1 Связанные списки

Напишите функцию (пусть онабудет называться reverse_linked_list), которая разворачивает связный список. На вход она принимает головную ноду, а на выход отдаёт хвостовую ноду исходного списка, но только теперь если попробовать пройтись по ней, она будет новой головной. С функцией из примера выше,

```python
class Node(object):
    def __init__(self, value=None, next_node=None):
        self.value = value
        self.next = next_node

    def __str__(self):
        return f"[Node with value {self.value}]"

def reverse_linked_list(head):
    if head is None or head.next is None:
        return head

    prev = None
    current = head
    next_node = None

    while current is not None:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev

def print_linked_list(head):
    cur = head
    while cur is not None:
        print(cur)
        cur = cur.next
```

```python
h, a, b, c, d = Node(1), Node(2), Node(3), Node("Внезапно"), Node(5)

h.next = a
a.next = b
b.next = c
c.next = d

print_linked_list(h)
print_linked_list(h)
h = reverse_linked_list(h)
print("---")
print_linked_list(h)
```

Задача 2 Пузырьки



```python
def bubble_sort_linked_list(head):
    if not head:
        return None

    def swap(node1, node2):
        node1.value, node2.value = node2.value, node1.value

    current = head
    while current:
        pointer = current.next
        while pointer:
            if current.value > pointer.value:
                swap(current, pointer)
            pointer = pointer.next
        current = current.next

    return head
```

```python
head = Node(3)
head.next = Node(2)
head.next.next = Node(4)
head.next.next.next = Node(1)
head.next.next.next.next = Node(5)

sorted_head = bubble_sort_linked_list(head)

print_linked_list(head)
```
